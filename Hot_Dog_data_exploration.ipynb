```python
import pandas as pd
df = pd.read_csv('data.csv')
display(df.head())
```

```python
df = df.dropna(axis=1)
display(df.head())
```

```python
df['Is hotdog a sandwich?'] = df['Is hotdog a sandwich?'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df= df.rename(columns={'What year are you in?': 'Year'})
display(df.head())
```

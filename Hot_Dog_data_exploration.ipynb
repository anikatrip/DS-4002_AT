```python
import pandas as pd
df = pd.read_csv('hot_dog.csv', encoding='latin-1')
display(df.head())
```

```python
df = df.dropna(axis=1)
display(df.head())
```

```python
df['Is hotdog a sandwich?'] = df['Is hotdog a sandwich?'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df= df.rename(columns={'What year are you in?': 'Year'})
display(df.head())
```

```python
df= df.rename(columns={'Is hotdog a sandwich?': 'Response'})
display(df.head())
```

```python
df= df.rename(columns={'ï»¿Timestamp': 'Time'})
display(df.head())
```

```python
display(df.dtypes)
```

```python
df['Time'] = df['Time'].astype(str)
df[['Date', 'Hour']] = df['Time'].str.split(' ', expand=True)
display(df.head())
```
